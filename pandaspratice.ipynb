```python
# sd
'''

1️⃣ Simple definition (1 line)

“Standard deviation measures how spread out the data is around the mean.”

2️⃣ How it happens (1 line)

“It calculates the average distance of each data point from the mean to show variability.”

3️⃣ Real-world examples (1 line each)
Scenario	Example (1 line)
Salaries	Mean = ₹50,000, SD = ₹5,000 → most earn 45,000–55,000
Exam scores	Mean = 70/100, SD = 15 → most students scored 55–85
Stock prices	Mean = ₹200, SD = ₹20 → most daily prices 180–220
Working hours	Mean = 40 hrs, SD = 5 → most employees work 35–45 hrs
✅ How to say in interview

“Standard deviation measures the spread of data around the mean. For example, in salaries, if the mean is ₹50,000 and SD is ₹5,000, most employees earn between ₹45,000 and ₹55,000. It helps us understand consistency, variation, and spot outliers.”

1️⃣ Identify lagging employees

By analyzing performance scores or KPIs:

Mean = average performance

SD = spread of performance

Employees below (Mean – 1 SD) → lagging / underperforming

HR can target training, mentoring, or support for these employees

2️⃣ Identify top performers

Employees above (Mean + 1 SD) → high performers

HR can reward, promote, or give incentives

3️⃣ Decide increments and raises

If performance is normally distributed:

Low SD → most employees perform similarly → uniform increments

High SD → wide variation → tailor increments based on individual performance

Example:

Performance scores: Mean = 70, SD = 10

Employee scoring 60 → below average → training / small raise

Employee scoring 85 → above average → higher increment / bonus

4️⃣ Other benefits

Detect anomalies → outliers may indicate errors or exceptional contributions

Plan workforce development → focus on lagging areas

Fairness in salary & appraisal → data-driven decisions

✅ Summary:

“Standard deviation helps HR see which employees are lagging, who are top performers, and how to distribute increments or training fairly based on actual performance data.”


'''
print(df['N'].std())
```

```python
# ============== Handling Missing values ====================================
'''
✅ Data_frame.dropna()

1-line definition:
Removes rows that contain missing (NaN) values but does not modify the original DataFrame.

Why we use it:
To clean rows that have incomplete data.

Code:

df2 = df.dropna()


Example Output (rows with NaN removed):
Before:

age   crop
25    rice
NaN   wheat
30    maize


After:

age   crop
25    rice
30    maize


Real-world use:
Remove crop records where temperature or rainfall data is missing.

✅ Data_frame.dropna(inplace=True)

1-line definition:
Permanently removes rows with missing values from the original DataFrame.

Why we use it:
To save memory and directly clean original data.

Code:

df.dropna(inplace=True)


Real-world use:
Clean drone sensor data by removing corrupted records.

✅ Data_frame.fillna(default_value)

1-line definition:
Replaces missing values with a specified value.

Why we use it:
To avoid losing data and keep dataset complete.

Code (fill with number):

df['temperature'] = df['temperature'].fillna(0)


Code (fill with text):

df['label'] = df['label'].fillna('unknown')


Example Output:
Before:

temperature
NaN
25.5


After:

temperature
0.0
25.5


Real-world use:
Fill missing crop labels with "unknown" instead of deleting row.

🔥 Interview One-Liner You Can Say

“I handle missing data using dropna() to remove incomplete rows or fillna() to replace missing values with appropriate defaults.”

Quick Summary Table
Method	Purpose
dropna()	Remove rows with missing values
dropna(inplace=True)	Remove permanently
fillna(value)	Replace missing values



1️⃣ Fill missing values with Mean

Definition: Replace NaN with average of the column.

Use case: Numeric data where the mean makes sense (e.g., salaries, heights).

df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print("After filling Salary with mean:\n", df)


Step-by-step:

Calculate mean: (5000+6000+5500+7000)/4 = 5875

Replace NaN with 5875

2️⃣ Fill missing values with Median

Definition: Replace NaN with median (middle value).

Use case: Numeric data with outliers (median is more robust than mean).

df['Age'].fillna(df['Age'].median(), inplace=True)
print("After filling Age with median:\n", df)


Step-by-step:

Age column (non-NaN): 25, 30, 28, 35 → median = 29

Replace NaN with 29

3️⃣ Fill missing values with Mode

Definition: Replace NaN with most frequent value.

Use case: Categorical data (department, city, gender).

df['Department'].fillna(df['Department'].mode()[0], inplace=True)
print("After filling Department with mode:\n", df)


Step-by-step:

Department values: HR, IT, IT, HR → mode = HR

Replace None with HR

4️⃣ Fill missing values with Max / Min

Definition: Replace NaN with maximum or minimum value of the column.

Use case: Extreme-case imputation, sometimes used in specific business rules.

# Fill NaN in Salary with maximum value
df['Salary'].fillna(df['Salary'].max(), inplace=True)

# Fill NaN in Age with minimum value
df['Age'].fillna(df['Age'].min(), inplace=True)

Final Cleaned DataFrame
   Salary   Age Department
0  5000.0  25.0        HR
1  6000.0  29.0        IT
2  5875.0  30.0        HR
3  5500.0  28.0        IT
4  7000.0  29.0        HR
5  5875.0  35.0        HR

✅ Summary Table
Method	Column Type	When to use	Example
mean	Numeric	Data is fairly symmetric	Salary
median	Numeric	Data has outliers	Age
mode	Categorical	Most common value	Department
max	Numeric	Business logic / extreme cases	Salary
min	Numeric	Business logic / extreme cases	Age


5️⃣ Fill with Standard Deviation (SD)

Definition: Standard deviation measures how spread out the data is.

Use case: Rarely used to fill missing values directly, but sometimes people fill with mean ± SD to simulate realistic variation.

Example: Fill NaN in Salary with Mean + SD
salary_mean = df['Salary'].mean()
salary_sd = df['Salary'].std()

# Fill missing values with mean + SD (example)
df['Salary'] = df['Salary'].fillna(salary_mean + salary_sd)


mean + SD adds a bit of variation rather than filling all with the exact mean.
'''


df2=df.dropna() # reove the duplicates values
df2
df.dropna(inplace=True) # reove the duplicates values permantaly
df
```
